```python
# Check Content
import os
from google.colab import userdata

DATASET_PATH = userdata.get('dataset_path')

for folder in os.listdir(DATASET_PATH):
    print(folder)
```

```python
# Check number of images per label
for label in os.listdir(DATASET_PATH):
    folder_path = os.path.join(DATASET_PATH, label)
    if os.path.isdir(folder_path):
        print(label, "has", len(os.listdir(folder_path)), "images")
```

```python
# Convert image to CSV
import cv2
import numpy as np
import pandas as pd

IMG_SIZE = 64  # resize all images to 64x64
rows = []
labels = []

for label in os.listdir(DATASET_PATH):
    folder = os.path.join(DATASET_PATH, label)
    if not os.path.isdir(folder):
        continue
    for filename in os.listdir(folder):
        filepath = os.path.join(folder, filename)
        img = cv2.imread(filepath)
        if img is None:
            continue
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        flat = img.flatten().tolist()
        rows.append(flat)
        labels.append(label)

df = pd.DataFrame(rows)
df['label'] = labels

df.to_csv("gesture_dataset.csv", index=False)
print("CSV saved with", len(df), "samples")
```
